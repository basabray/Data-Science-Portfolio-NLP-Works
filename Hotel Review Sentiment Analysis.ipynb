# Hotel Review Sentiment Analysis: 
- Performed the sentoment analysis using vader setiment scoring technique

### VaderSentiment technique:

```python
import pandas as pd
import nltk
from gensim.models import Word2Vec
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('train.csv')
df = data[:1000]
```

```python
df.shape
```

```python
from keras.preprocessing.text import text_to_word_sequence
#Applying word level tokenization and stroing into a column named, 'word_tokens'
word_tokens = []
for i in df['Description']:
  word_tokens.append(text_to_word_sequence(i))
df['word_tokens'] = word_tokens
df.head()
```

```python
#Stop words combining step:

#get all the stop words from nltk library:
import nltk
nltk.download('stopwords')
stop = nltk.corpus.stopwords.words('english')

#get all the stop words from sklearn library:
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS as sklearn_stop

#now combine the stop words
combined_stop = list(set(stop).union(sklearn_stop))


#Stop words removal step: After removing stop wird storing the data into a list named, 'clean data'
clean_data = []

for sen in df['word_tokens']:
    for words in sen:
        if words in combined_stop:
            sen.remove(words)
    clean_data.append(sen)

#print only the 1st record after removing the stop words
clean_data[0]   
```

```python
#Creting the column named, 'vader_sentiment' for each review 
score = []
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sa = SentimentIntensityAnalyzer()
for doc in clean_data:
    for review in doc:
        scores = sa.polarity_scores(review)
    score.append(scores)
df['vader_sentiment'] = score


#Storing only the compound score to a column named, 'compound_scores'
compound_scores = []

for i in df.vader_sentiment:
    for key,value in i.items():
        if key=='compound':
            compound_scores.append(value)
df['compound_scores'] = compound_scores


#printing the firts 30 reviews with their vader sentiment score details
df[:30]
```

```python
print(df[df['compound_scores']>0.0000].shape[0])      #Positive  reviews
print(df[df['compound_scores']<0.0000].shape[0])      #Negative reviews
print(df[df['compound_scores']==0.0000].shape[0])     #Neutral reviews
```

# Hotel Review Sentiment Prediction: 
### TF-IDF model to vectorize the words and then applied Logistic regression to classify the reviews:
##### Hotel review 'Description' is the input avraible and the target variable is 'Is_Response'

```python
df2 = df[['Description','Is_Response']]
df2.head()
```

```python
df2.shape[0]
```

###### Text data pre-processing:

```python
#clean the data: remove special charc, punctuation etc
import re
import string

#This function converts to lower-case, removes square bracket, removes numbers and punctuation
def text_clean_1(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

cleaned1 = lambda x: text_clean_1(x)


#store the cleaned data
df2['cleaned_description'] = pd.DataFrame(df2.Description.apply(cleaned1))
df2.head(10)
```

#### Train-test split:

```python
from sklearn.model_selection import train_test_split

x = df2['cleaned_description']
y = df2['Is_Response']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 225)

print('Independent Variable train :', len(x_train))
print('Independent Variable test:', len(x_test))
print('Dependent Variable train :', len(y_train))
print('Dependent Variable test  :', len(y_test))
```

#### Model building:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

tvec = TfidfVectorizer()
clf2 = LogisticRegression(solver = "lbfgs")
```

```python
from sklearn.pipeline import Pipeline  
#pipeline will execute the steps 1 after another. At first, word vectors will be created using TF-IDF
#and then Logistic reg will be applied to classify the data based on the review descriptions

model = Pipeline([('vectorizer',tvec),('classifier',clf2)])     #First apply vectorizer and then classifier
model.fit(x_train, y_train)


from sklearn.metrics import confusion_matrix
predictions = model.predict(x_test)
confusion_matrix(predictions, y_test)
```

##### Model Evaluation:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print("Accuracy : ", accuracy_score(predictions, y_test))
print("Precision : ", precision_score(predictions, y_test, average = 'weighted'))
print("Recall : ", recall_score(predictions, y_test, average = 'weighted'))
```
